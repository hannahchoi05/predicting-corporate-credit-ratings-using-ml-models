```python
import pandas as pd

capital_iq_ratings = pd.read_csv('ratings_C.csv')
compustat_fundamentals = pd.read_csv('ratio_C.csv')
wrds_gvkey_link = pd.read_csv('link_C.csv')  # The linking table from WRDS

merged_step1 = capital_iq_ratings.merge(
    wrds_gvkey_link,
    left_on='Company_id',
    right_on='companyid',
    how='inner'
)

final_data = merged_step1.merge(
    compustat_fundamentals,
    on='gvkey',
    how='inner'
)

final_data.to_csv('linked_ratings_fundamentals.csv', index=False)
```
